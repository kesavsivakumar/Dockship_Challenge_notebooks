```


from keras.applications import ResNet50,VGG19
from keras.models import Model, Sequential
from keras import layers
# load model without classifier layers
model=Sequential()
model.add(VGG19(input_shape=(256,256,3),include_top=False,pooling='avg',weights='imagenet'))
#print(model._get_trainable_state())
# Set every layer to be non-trainable:
#for k,v in model._get_trainable_state().items():
    #k.trainable = False
#model.add(Flatten())
model.add(layers.Dense(1000, activation='relu', kernel_initializer='he_uniform'))
model.add(layers.Dense(128, activation='relu', kernel_initializer='he_uniform'))
#model.add(Dense(,activation='relu'))
model.add(layers.Dense(50,activation='softmax'))
# summarize
model.layers[0].trainable = False
model.summary()

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```
from keras.preprocessing import image as image_utils
from keras.applications.imagenet_utils import decode_predictions
from keras.applications.imagenet_utils import preprocess_input
from keras.applications import VGG16
import numpy as np
from numpy import array
from numpy import argmax
from keras.utils import to_categorical
from keras.callbacks import EarlyStopping,ModelCheckpoint
```

```
trainY=np.array(trainY)
encoded_y=to_categorical(trainY)
```

```

from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(trainX, encoded_y, test_size=0.05, random_state=42)
```

```
es = EarlyStopping(monitor='loss', mode='min', verbose=1,patience=64)
mc = ModelCheckpoint(r'BC.h5', monitor='accuracy',mode='max', verbose=1, save_best_only=True)
 # fit model
history = model.fit( X_train,y_train,validation_data=(X_val,y_val),batch_size=128,epochs=100, verbose=1,callbacks=[es,mc])
#history = model.fit( trainX,encoded_y,batch_size=128,epochs=100, verbose=1,callbacks=[es,mc])
```

```
import cv2
import re
import os
import pandas as pd
import numpy as np
import pytesseract
```

```
result=[]
```

```
count=1
for path in os.listdir('TextRecognition/TEST'):
  img=cv2.imread('TextRecognition/TEST/'+path)
  #print(img.shape)
  #img=cv2.resize(img,(256,256))
  #print(img.shape)
  gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
  gray, img_bin = cv2.threshold(gray,128,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)
  gray = cv2.bitwise_not(img_bin)
  kernel = np.ones((2, 1), np.uint8)
  img = cv2.erode(gray, kernel, iterations=1)
  img = cv2.dilate(img, kernel, iterations=1)
  out_below = pytesseract.image_to_string(img)
  out_below=out_below.upper()
  ansi =re.compile(r'(\x9B|\x1B\[)[0-?]*[ -\/]*[@-~]')
  out_below=ansi.sub('', out_below)
  result.append([path,out_below])
  count=count+1
  print(count)
```

```
result
```

```
escapes = ''.join([chr(char) for char in range(1, 32)])
for i in range(len(result)):
  #result[i][1]=re.sub(r'[\x9B|\x1B]','',result[i][1])
  result[i][1] = result[i][1].translate( escapes)
```

```
for i in range(len(result)):
  result[i][1]=re.sub(r'\x0b','',result[i][1])
  #result[i][1] = result[i][1].translate( escapes)
```

```
for i in range(len(result)):
  result[i][1]=re.sub(r'\r','',result[i][1])
  #result[i][1] = result[i][1].translate( escapes)
```

```
for i in range(len(result)):
  result[i][1]=re.sub(r' ','',result[i][1])
  #result[i][1] = result[i][1].translate( escapes)
```

```
for i in range(len(result)):
  result[i][1]=re.sub(r'[:+-/*@()?<>;]','',result[i][1])
  #result[i][1] = result[i][1].translate( escapes)
```
